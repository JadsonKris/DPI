```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_rgb(image):
    colormap1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(colormap1)
    plt.show()
    chans=cv2.split(colormap1)
    colors = ("r", "g", "b")
    channels = ("RED", "GREEN", "BLUE")
    plt.title("Color histogram RGB")
    plt.xlabel("Bins")
    plt.ylabel("Number of pixels")
    
    for (chan, c) in zip(chans, colors):
        hist=cv2.calcHist([chan], [0], None, [256], [0,256])
        plt.plot(hist, color=c)
        plt.xlim([0,256])
    
    plt.legend(channels)
    plt.show()
    
    #fig, (ax1, ax2) = plt.subplots(1, 2)
    #ax1.plt()
    
```

```python
def plot_hsv(image):
    colormap = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    plt.imshow(colormap)
    plt.show()
    chans=cv2.split(colormap)
    colors=("r", "g", "b")
    channels=("HUE", "SATURATION", "VALUE")
    plt.title("Color histogram HSV")
    plt.xlabel("Bins")
    plt.ylabel("Number of pixels")
    
    for (chan, c) in zip(chans, colors):
        hist=cv2.calcHist([chan], [0], None, [256], [0,256])
        plt.plot(hist, color=c)
        plt.xlim([0,256])
    
    plt.legend(channels)
    plt.show()
```

```python
def plot_YCrCb(image):
    colormap = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
    plt.imshow(colormap)
    plt.show()
    chans=cv2.split(colormap)
    channels=("LUMA", "CR", "CB")
    colors =("y", "r", "b")
    plt.title("Color histogram YCrCb")
    plt.xlabel("Bins")
    plt.ylabel("Number of pixels")
    
    for (chan, c) in zip(chans, colors):
        hist=cv2.calcHist([chan], [0], None, [256], [0,256])
        plt.plot(hist, color=c)
        plt.xlim([0,256])
    
    plt.legend(channels)
    plt.show()
```

```python
def plot_histo(path):
    image = cv2.imread(path)
    #RGB
    plot_rgb(image)
    plot_hsv(image)
    plot_YCrCb(image)
```

```python
plot_histo("baboon.png")
```

```python
plot_histo("chips.png")
```
