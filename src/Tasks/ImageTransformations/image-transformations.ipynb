```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.misc
from scipy import ndimage
```

```python
#Crop and flip an image using Numpy array indexing.
face = scipy.misc.face(gray=True)
lx, ly = face.shape
crop_face = face[lx//4:-lx//4, ly//4:-ly//4]
flip_ud_face = np.flipud(face)

plt.subplot(151)    
plt.imshow(face, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(152)
plt.imshow(crop_face, cmap=plt.cm.gray)
plt.title('crop image')
plt.axis('off')

plt.subplot(153)
plt.imshow(flip_ud_face, cmap=plt.cm.gray)
plt.title('flip image')
plt.axis('off')

plt.subplots_adjust(wspace=0.02, hspace=0.3, top=1, bottom=0.1, left=0,
                    right=1)

plt.show()
```

```python
#Implement image translation using Numpy and OpenCV.

image = cv2.imread("images/faces.jpg")
height, width = image.shape[:2]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()

quarter_height, quarter_width = height / 4, width / 4

T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

img_translation = cv2.warpAffine(image, T, (width, height))

plt.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB))
plt.title("Translation")
plt.show()
```

```python
#Implement image translation using Numpy and OpenCV.
#numpy
image = cv2.imread("images/faces.jpg")


height, width = image.shape[:-1]
shift = 100
rolled = np.roll(image, shift, axis=[0, 1])
rolled = cv2.rectangle(rolled, (0, 0), (width, shift), 0, -1)
rolled = cv2.rectangle(rolled, (0, 0), (shift, height), 0, -1)
plt.imshow(rolled)
```

```python
# Implement image rotation using Numpy and OpenCV.
#numpy
face = scipy.misc.face(gray=True)

rotate_face = ndimage.rotate(face, 45)

plt.figure(figsize=(12.5, 2.5))


plt.subplot(151)
plt.imshow(face, cmap=plt.cm.gray)
plt.title('Original')
plt.axis('off')
plt.subplot(152)
plt.imshow(rotate_face, cmap=plt.cm.gray)
plt.title('Rotate')
plt.axis('off')

plt.subplots_adjust(wspace=0.02, hspace=0.3, top=1, bottom=0.1, left=0,
                    right=1)

plt.show()
```

```python
#Implement image rotation using Numpy and OpenCV.
#OpenCV
face = cv2.imread("images/faces.jpg")
image = cv2.rotate(face, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
#Implement image resizing using OpenCV

image = cv2.imread("images/faces.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))

stretch_near = cv2.resize(image, (780, 540), interpolation = cv2.INTER_NEAREST)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```

```python
# Implement bitwise operations: AND, OR, XOR.

rectangle = np.zeros((300, 300), dtype="uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1)
plt.imshow(cv2.cvtColor(rectangle, cv2.COLOR_BGR2RGB))
plt.title("Rectangle")
plt.show()


circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, 255, -1)
plt.imshow(cv2.cvtColor(circle, cv2.COLOR_BGR2RGB))
plt.title("Circle")
plt.show()
    
    
bitwise_and = cv2.bitwise_and(rectangle, circle)
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
plt.title("bitwise AND")
plt.show()


bitwise_or = cv2.bitwise_or(rectangle, circle)
plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB))
plt.title("bitwise OR")
plt.show()


bitwise_xor = cv2.bitwise_xor(rectangle, circle)
plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB))
plt.title("bitwise XOR")
plt.show()
```

```python
image = cv2.imread("images/faces.jpg")

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.show()

mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.circle(mask, (300, 150), 100, 255, -1)
masked = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.title("Mask")
plt.show()

plt.imshow(cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))
plt.title("Mask Applied to Image")
plt.show()
```
